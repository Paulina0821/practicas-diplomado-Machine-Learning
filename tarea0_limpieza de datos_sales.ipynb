```python
#importar librerias que usaremos 
import numpy as np 
import seaborn as sns
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px

sales = pd.read_csv("sales_record.csv", sep=",")
#cargamos los datos que vamos a usar 
print(sales.shape)# para ver el numero de columnas y filas 
print(sales.head()) #para ver los primeros elementos del dataframe 

sales.head()
```

```python
salescop =  sales1.copy()#copia para no modificar el archivo original
salescop 
```

```python
salescop['Order Date'] = pd.to_datetime(salescop['Order Date'], format='mixed')
                                        
#pd.to_datetime(o_retail['InvoiceDate'],format='%d/%m/%Y')#comando general para cambiar a formato de fecha
salescop.info()#vemos que al transformar a formato de fecha perdimos datos
```


## 1. Promedio de unidades vendidas por región y país


```python
salescop.Region.unique()
```

```python
dff= salescop.groupby(['Region', 'Country'])['Units Sold'].mean().sort_values().reset_index()

dff
```

```python
px.bar(data_frame= dff, x="Country", y="Units Sold", color= "Region")
```

```python
sns.set(rc={'figure.figsize':(14,30)})
sns.set_style("whitegrid")

sns.barplot(data=dff, y="Country", x="Units Sold", hue="Region", orient="h")
```

## 2. Promedio de ventas por tipo de Item y por región


```python
df1= salescop.groupby(['Region', 'Item Type'])['Units Sold'].mean().reset_index()

df1
```

```python
#con esta grafica vemos el tipo de producto que se compra por region
sns.set(rc={'figure.figsize':(14,30)})
sns.set_style("whitegrid")

sns.barplot(data=df1, y="Item Type", x="Units Sold", hue="Region", orient="h")
```

## 3. Promedio de Beneficio total por país


```python
df2= salescop.groupby(['Region', 'Country'])['Total Profit'].mean().reset_index()

df2
```

```python
#con esta grafica se muestra el promedio del benedicio total por Region y País
sns.set(rc={'figure.figsize':(14,30)}) 
sns.set_style("whitegrid")

sns.barplot(data=df2, y="Country", x="Total Profit", hue="Region", orient="h")
```

```python
df3= salescop.groupby(['Region', 'Country'])['Total Profit'].mean().sort_values().reset_index()

df3
```

```python
#con esta grafica se muestra el promedio de "beneficio total" por país, aquí es muy claro observar que el promedio menor de 
#beneficio total corresponde a Kirguistán y el mayor promedio a Hungria
sns.set(rc={'figure.figsize':(14,30)})
sns.set_style("whitegrid")

sns.barplot(data=df3, y="Country", x="Total Profit", hue="Region", orient="h")
```

## 4. Promedio de Beneficio total Por fecha de venta y region



```python
salescop= salescop.rename(columns={'Order Date': 'Order_Date'})#remombramos columna de la fecha
```

```python
#con estos comandos voy a separra la fecha y el mes
salescop["Year"] = pd.DatetimeIndex(salescop.Order_Date).year
salescop[:3]

salescop["Month"] = pd.DatetimeIndex(salescop.Order_Date).month
salescop[:3]
```

```python
df4= salescop.groupby(['Region', 'Year'])['Total Profit'].mean().sort_values().reset_index()#beneficio total por año de venta y por region

df4
```

```python
#grafico que muestra el beneficio total por region y por año
sns.set(rc={'figure.figsize':(14,30)})
sns.set_style("whitegrid")

sns.barplot(data=df4, y="Region", x="Total Profit", hue="Year", orient="h")
```

## 5. Promedio de ganancia total por año 

```python
df5= salescop.groupby('Year')['Total Profit'].mean().reset_index()#promedio beneficio total por año

df5
```

```python
#vemos que el año con menor beneficio es 2016 y con mayor beneficio total es 2012
sns.set(rc={'figure.figsize':(6,10)})
sns.set_style("whitegrid")

sns.barplot(data=df5, y="Year", x="Total Profit", orient="h")
```

```python
df6= salescop.groupby('Month')['Total Profit'].mean().reset_index()#

df6
```

```python
#en la grafica vemos que agosto, septiembre y noviembre son los meses con menor beneficio total, mientras que enero y diciembre tienen los mayores promedios de beneficio total
sns.set(rc={'figure.figsize':(6,10)})
sns.set_style("whitegrid")

sns.barplot(data=df6, y="Month", x="Total Profit", orient="h")
```

```python
df7= salescop.groupby(['Region', 'Month'])['Units Sold'].mean().reset_index()#promedio de beneficio total por mes de venta y por region

df7
```

```python
#grafico que muestra las unidades vendidas por region y por mes de venta 
sns.set(rc={'figure.figsize':(10,20)})
sns.set_style("whitegrid")

sns.barplot(data=df7, y="Region", x="Units Sold", hue="Month", orient="h")
```
