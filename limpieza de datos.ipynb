```python
#importar librerias que usaremos 
import numpy as np 
import seaborn as sns
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("weblog-iies.csv", sep=",")
#cargamos los datos que vamos a usar 
print(data.shape)# para ver el numero de columnas y filas 
print(data.head()) #para ver los primeros elementos del dataframe 

data

#hay 11 columnas y 9110 filas
```

```python
# hay una columna sin encabezado, pero sabemos que es el tiempo que tarda en recibir respuesta, se le dará el nombre de "tiempo"
data = data.rename(columns={'Unnamed: 4': 'Tiempo'})
print(data)
```

# Trabajando con el DataFrame

ya que las columnas tienen nombres podemos crear un DataFrame para trabajar los valores nulos y posterior a ello se hará la trasnformación de datos 

```python
datos=pd.DataFrame(data)
```

```python
datos.dtypes#vemos el tipo de dato que hay
```

```python
datos.info()#vemos que hay valores nulos y el tipo de dato no es correcto en algunas columnas
```

```python

#primero haré una copia de los datos para no perder el archivo original
datos1 =  datos.copy()
datos1 
```

```python
datos1.isna().sum()# vemos el total de datos nulos por columna, considerando que tenemos 9110 registros podemos considerar
#rellenar los valores nulos en algunas columnas y eliminar en otras 
```

```python
datos1['Address']= datos1['Address'] .fillna('Unknown')#podemos llenar los datos faltantes en la dirección
#IP con "Unknown)
```

```python
datos1.isna().sum()#vemos que ya fueron llenados los especios vacíos
```

```python
#ahora vamos con la columna SSL, queremos ver que datos hay para saber si podemos rellenar o eliminar datos faltantes
datos1["SSL"].unique()
#con esto vemos que el unico dato es TLSv1.2 y - que tambien puede indicar que no hay registro, por lo cual para asegurarme
#de que estoy llenando los espacios con el valor adecuado voy a obtener la moda en esta columna
```

```python
datos1["SSL"].mode()#con este resultado, voy a rellenar los datos faltantes con esta moda
```

```python
datos1['SSL']= datos1['SSL'] .fillna('TLSv1.2')
```

```python
datos1["SSL"].unique()
```

```python
datos1.isna().sum()#ya no hay datos faltantes en SSL
```

```python
##seguire con la columna Bytes, y en este caso quise rellenar datos nulos con la mediana de los datos 
#usando el siguiente codigo #print(datos1['Bytes'].median(skipna= True )) sin embargo, marca error ya que el tipo de dato no es numerico, así que
#lo siguiente es intentar hacer la transformacion

datos1['Bytes'] = pd.to_numeric(datos1['Bytes'], errors='coerce')#con este codigo converti la columna de Bytes 
#a numerico
```

```python
datos1.dtypes#vemos el tipo de dato que hay y vemos que Bytes ya es numerico flotante y aun tenemos datos nan
```

```python
print(datos1['Bytes'].median(skipna= True )) #ahora si obtenemos la mediana (sin tomar en cuenta los valores faltantes)
#y con este valor reemplazo todos los  nan de esta columna 
```

```python
datos1['Bytes']= datos1['Bytes'].fillna(21104.0)
```

```python
datos1.dtypes #vemos el tipo de dato que hay
```

```python
datos1.isna().sum()#revisamos que ya no hay valores faltantes en Bytes
```

```python
#seguimos con la version
datos1["Version"].unique()

#en este caso decido rellenar datos con bfill

```

```python

datos1["Version"] = datos1["Version"].replace("", None).bfill()
```

```python
datos1.isna().sum()#revisamos que ya no hay valores faltantes en version
```

```python
datos1 = datos1.replace(['-"'],[" "])#quitare lo guiones tambien de esta columna
```

```python
datos1 = datos1.drop(datos1[datos1['Type']==" "].index)
```

```python
datos1.Type.unique()
```

```python
datos1=datos1.dropna(subset=["Criptografia", "Type","URL"])#finalmente eliminare los datos faltantes de las
#columnas criptografia,type y url
```

```python
datos1#teminamos con 8929 registros y 11 columnas
```

```python
datos1.isna().sum()#vemos que el dataframe ya esta limpio
```

```python
datos1
```

```python
#falta transformar las fechas al formrato correcto 
datos1.info()
```

```python
datos1['Date'] = pd.to_datetime(datos1['Date'])
#datos1['Date'] = pd.to_datetime(datos1['Date'],format='%d%m%Y',  errors='ignore')
```

```python

datos1["Time"]=pd.to_timedelta('00:' + datos1["Time"]) 
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
datos1['Type'].value_counts().head(40).plot.bar(color = "cyan")
plt.title('Most Popular Type by the Users', fontsize = 20)
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
datos1['URL'].value_counts().head(40).plot.bar(color = color)
plt.title('Most Popular URLs by the Users', fontsize = 20)
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
data['Address'].value_counts().head(40).plot.bar(color = 'cyan')
plt.title('Most Popular Address of Logins', fontsize = 20)
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
datos1['Version'].value_counts().head(40).plot.bar(color = "cyan")
plt.title('Most Popular Version by the Users', fontsize = 20)
```

```python

datos1.groupby(["Address", "URL"],  as_index=False)["Bytes"].agg("mean", numeric_only=True)
```

```python
datos1.groupby('Address',as_index=False)[['URL', 'Bytes']].sum()
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
datos1.groupby('Address')[['URL', 'Bytes']].agg("mean", numeric_only=True).head(40).plot.bar(color = 'cyan')
plt.title('Most Popular Address of Logins', fontsize = 20)
```

```python
datos1.groupby("Address")["Bytes"].mean()
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
datos1.groupby('Address')[['URL', 'Bytes']].sum().head(40).plot.bar(color = 'cyan')
plt.title('Cantidad de Bytes enviados respecto a la URL', fontsize = 20)
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
datos1.groupby("Address")["Bytes"].mean().head(40).plot.bar(color = 'cyan')
plt.title('Most Popular Address and Bytes of Logins', fontsize = 20)
```
